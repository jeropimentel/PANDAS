```python
ser = pd.Series([2,5,7,4], index=['one','two','three','four']) # reindex
ser
```

```python
ser.reindex(['two','four','five','one']) #reindex modifica el orden, deletea o agregar rows, si pones indices random agrega NaN
```

```python
ser3 = pd.Series([1,5,6,3],index=[0,3,5,6])
ser3
```

```python
ser3.reindex(range(6), method='ffill')#As you can see from the result, the indexes that were not present in the original series
                                      #were added. By interpolation, those with the lowest index in the original series
```

```python
ser3.reindex(range(6), method='bfill')
```

```python
# dropping values
ser = pd.Series(np.arange(4.), index=['red','blue','yellow','white'])
ser
```

```python
ser.drop(['yellow','blue'])
```

```python
frame = pd.DataFrame(np.arange(16).reshape((4,4)), index=['red','blue','yellow','white'],columns=['ball','pen','pencil','paper'])
frame
```

```python
frame.drop(['blue','yellow']) # por default elige rows
```

```python
frame.drop(['ball','pen'], axis=1) # axis 1 equivale a columna
```

```python
# align index between df
s1 = pd.Series([3,2,5,1],['white','yellow','green','blue'])
s2 = pd.Series([1,4,7,2,1],['white','yellow','black','blue','brown'])

s1 + s2
```

```python
frame1 = pd.DataFrame(np.arange(16).reshape((4,4)),index=['red','blue','yellow','white'],columns=['ball','pen','pencil','paper'])
frame2 = pd.DataFrame(np.arange(12).reshape((4,3)), index=['blue','green','white','yellow'], columns=['mug','pen','ball'])

frame1 + frame2 #  al sumarlos quedan por fuera los valores que no compartes valor de columna y row, peras con peras , manzanas con manzanas
```
