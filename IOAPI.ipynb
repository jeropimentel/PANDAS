```python
#convertir el tree en un dataframe
def etree2df(root):
    column_names = []
    for i in range(0,len(root.getchildren()[0].getchildren())):
        column_names.append(root.getchildren()[0].getchildren()[i].tag)
    frame = pd.DataFrame(columns = column_names)
    
    for j in range(0, len(root.getchildren())):
        obj = root.getchildren()[j].getchildren() # todos los elementos en la posicion j  de getchildren
        texts= []
        for k in range(0, len(column_names)):
            texts.append(obj[k].text) #agrega a texts cada valor k presente en obj
        row = dict(zip(column_names, texts)) # crea un dicionario con los valores con las columnas
        row_s = pd.Series(row)
        row_s.name = j # numeros de rows, sirve de índice para el pd.Series
        frame = frame.append(row_s) # agrega la serie al data frame
    return frame
            
```

```python
#usualy json files do not have a tabular form so is necesary to normalize- to convert the dict structure into tabular
from pandas.io.json import json_normalize
import json
file = open('books1.json', 'r')
text = file.read()
text = json.loads(text)
text
```

```python
json_normalize(text,'books') # va a leer y pasar a todos los elementos que tengan el key books
```

```python
json_normalize(text,'books',['writer', 'nationality']) # devuelve todos los que están debajo del key books y le suma de writer y nat
```
