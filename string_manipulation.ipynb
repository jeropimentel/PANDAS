```python
# para evitar ese espacio que queda se pueden usar las funciones split() y strip() corta en espacios
tokens = [s.strip() for s in text.split(',')]
tokens
```

```python
# también puedo pasar el mismo for-loop pero nombrando cada elemnto
adress, city = [s.strip() for s in text.split(',')]
city
```

```python
# para concatenar se puede utilizar el '+'
adress + ', ' + city
```

```python
# REGeX
import re
"""
pattern matching
substitution
splittin

"""
```

```python
text = "This is an\t odd \n text!"
text
```

```python
# the form to split a seq with one or more separator char is \+s
re.split('\s+', text)
```

```python
# findall - devuelve un patrón encontrado en un texto. Por ejemplo A\w+ busca las A y todas las palabras que le siguen
text = 'This is my address: 16 Bolton Avenue, Boston'
re.findall('A\w+', text)
```

```python
re.findall('[A,a]\w+', text) # el corchete A,a busca cualquier A o a seguida por un texto
```

```python
# besides findall .search() returns the first values
search = re.search('[A,a]\w+', text)
search
```

```python
# para conocer el valor hay que usar el span que indica donde empieza y donde termina
start = search.start()
end = search.end()
search = text[start:end]
search
```
