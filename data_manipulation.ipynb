```python
frame = pd.DataFrame({ 'item':['ball','mug','pen','pencil','ashtray'],'color':['white','red','green','black','yellow']})

# si se quiere pasar otra columna con nuevos valores se arma un diccionario que tenga keys en comun con una de la columnas
prices = { 'ball' : 5.56,
'mug' : 4.20,
'bottle' : 1.30,
'scissors' : 3.41,
'pen' : 1.30,
'pencil' : 0.56,
'ashtray' : 2.75
         }
frame['price'] = frame['item'].map(prices) # el map acepta una función o un diccionario que mapear
frame
```

```python
re_column= {
    'item': 'object',
    'price': 'value' 
}

frame.rename(columns=re_column, index=new_index)
```

```python
# rename crea un nuevo df dejando el original unchanged, para cambiar el objeto original se coloca inplace= True
frame.rename(columns={'item':'object'}, inplace=True)
frame
```

```python
# discret and bining
results = [12,34,67,55,28,90,99,12,3,56,74,44,87,23,49,89,87]
bins = [0, 25, 50, 75, 100] # queremos separar el rango de valores en 4

cat = pd.cut(results, bins)
cat # transforma los valores en un rango de bins
```

```python
cat.categories # devuelve los tipos de bins
```

```python
cat.codes # asigna un código a cada bin y muestra cuántas veces aparece
```

```python
# para saber lo ocurrencia de c/u
pd.value_counts(cat)
```

```python
bin_names = ['unlikely','less likely','likely','highly likely']
cat = pd.cut(results, bins, labels=bin_names) # asigna nombres a los bins
pd.value_counts(cat)
```

```python
# también se puede cortar en bins de un dado tamanio
pd.cut(results, 5).value_counts()
```

```python
# para dividir la muestra en quintilos, decilos, etc qcut(), siendo el segundo término la cantidad de divisiones
quintiles = pd.qcut(results, 5)
quintiles.value_counts()
```

```python
# DATA AGGREGATION
'''
ejemplo son count(), sum(), mean() - hacen operaciones que afectan a toda la matriz y devuelve un nro no una matriz-
group_by():
         .splitting() -divisions into groups 
         .applying() - app de una funcion en cada grupo
         .combining() - combinar de todos los resultados por diferentes grupos
    
'''
frame = pd.DataFrame({ 'color': ['white','red','green','red','green'],'object': ['pen','pencil','pencil','ashtray','pen'],'price1' : [5.56,4.20,1.30,0.56,2.75],'price2' : [4.75,4.12,1.60,0.75,3.15]})
frame
```

```python
# si queremos agrupar los price1 y usar los color como label podos usar el groupby
group = frame['price1'].groupby(frame['color'])
group # genera un gropu object
```

```python
group.mean() # devuelve el promedio del precio1 para cada color
```

```python
group.sum()
```

```python
# se puede establecer una jerarquía de agrupamiento usando disintas columnas como agrupadores
ggroup = frame['price1'].groupby([frame['color'], frame['object']])
ggroup.mean()
```

```python
ggroup.sum()
```

```python
frame[['price1','price2']].groupby(frame['color']).mean()  # también se puede llamar todas col valores de una y realizar el calculo aggreado de 1
```

```python
frame.groupby(frame['color']).mean() # sin aclarar las col de frame por default toma todas las col con nros
```

```python
frame.groupby('color').mean()
```

```python
# iteration over groups - quiero conocer los nombres y grupos
for name, groups in frame.groupby('color'):
    print(name) # título del grupo
    print(groups) # el grupo en cuestión
```

```python
# para hacerlo más legible podemos agregad el .add_prefix()
means = frame.groupby('color').mean().add_prefix('mean_')
means
```

```python
# functions on groups
group = frame.groupby('color')
group['price1'].quantile(0.6)
```

```python
def range(series):
    return series.max() - series.min()

group['price1'].agg(range)
```

```python
group.agg(range)
```

```python
# también se pueden colocar un array de funciones que se convierten en las nuevas colmnas
group['price1'].agg(['mean','std',range]).fillna(0) # built in functions pasan como string
```

```python
# advanced data aggregation
frame = pd.DataFrame({ 'color':['white','red','green','red','green'],'price1':[5.56,4.20,1.30,0.56,2.75],'price2':[4.75,4.12,1.60,0.75,3.15]})
frame
```
