```python
serd = pd.Series([1,0,2,1,2,3], index=['white','white','blue','green',' green','yellow'])
serd
```

```python
serd.unique() # funcion para seleccionar valores únicos
```

```python
serd.value_counts() # selecciona valores únicos y cuenta la ocurrencia de c/u
```

```python
serd.isin([0,3]) # func busca en la serie los valores 0 y 2
```

```python
serd[serd.isin([0,3])] # selecciona solamente los valores que son True- devuelve valor e index
```

```python
#Nan Values (Not a Number)- isnull() notnull()- identifican valores sin valor
s2 = pd.Series([5,-3,np.NaN,14])
s2.isnull() # devuelve los null values
```

```python
s2.notnull() # devuelve los notnull values
```

```python
s2[s2.notnull()] # devuelve los index y values de los notnull
```

```python
s2[s2.isnull()]
```

```python
# create a series from a dictionary
mydict = {'red': 2000, 'blue': 1000, 'yellow': 500, 'orange': 1000} # key funciona como index 
myseries = pd.Series(mydict)
myseries
```

```python
# para cambiar los valores de los índices se puede agregar un index al series
colors=['red','yellow','orange','blue','green']

myseries = pd.Series(mydict, index=colors)
myseries
```

```python
#operaciones entre series
mydict2 = {'red':400,'yellow':1000,'black':700}
myseries2 = pd.Series(mydict2)
myseries3 = myseries + myseries2 # devuelve la suma de los valores que puede sumar, pero puede sumar red and yellow
```

```python
myseries3[myseries3.notnull()]
```
